```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from keras_convattention.convattention import ConvAttentionLayer
from keras_convattention.attention import Attention
# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, header=0, index_col=0, parse_dates=True)

# Preprocessing the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))

# Function to create the dataset in the format (X, y)
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Create the dataset with a time_step of 12 (predict the next month based on the last 12 months)
time_step = 12
X, y = create_dataset(scaled_data, time_step)

# Reshape X to be 3D [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split the data into training, test, and validation sets (80% train, 10% validation, 10% test)
train_size = int(len(X) * 0.8)
val_size = int(len(X) * 0.1)
X_train, X_val, X_test = X[:train_size], X[train_size:train_size+val_size], X[train_size+val_size:]
y_train, y_val, y_test = y[:train_size], y[train_size:train_size+val_size], y[train_size+val_size:]
```

```python
# Model 1: LSTM + Attention Model
def build_lstm_attention_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Attention(units=128, score='bahdanau'))  # Using the custom Attention layer
    model.add(Dense(1))  # Output layer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Model 2: LSTM + CNN + Attention Model
def build_lstm_cnn_attention_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(ConvAttentionLayer(filters=64, kernel_size=3, pool_size=2, units=128, score='bahdanau', input_shape=input_shape))  # CNN + Attention layer
    model.add(Dense(1))  # Output layer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train models
lstm_attention_model = build_lstm_attention_model((X_train.shape[1], X_train.shape[2]))
lstm_cnn_attention_model = build_lstm_cnn_attention_model((X_train.shape[1], X_train.shape[2]))

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

lstm_attention_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val), verbose=1, callbacks=[early_stopping])
lstm_cnn_attention_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val), verbose=1, callbacks=[early_stopping])

# Predict using both models
lstm_attention_preds = lstm_attention_model.predict(X_test)
lstm_cnn_attention_preds = lstm_cnn_attention_model.predict(X_test)

# Inverse transform predictions and actual values to get them back to original scale
lstm_attention_preds = scaler.inverse_transform(lstm_attention_preds)
lstm_cnn_attention_preds = scaler.inverse_transform(lstm_cnn_attention_preds)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate Mean Squared Error, Mean Absolute Error, and R2 score for both models
lstm_attention_mse = mean_squared_error(y_test_actual, lstm_attention_preds)
lstm_cnn_attention_mse = mean_squared_error(y_test_actual, lstm_cnn_attention_preds)

lstm_attention_mae = mean_absolute_error(y_test_actual, lstm_attention_preds)
lstm_cnn_attention_mae = mean_absolute_error(y_test_actual, lstm_cnn_attention_preds)

lstm_attention_r2 = r2_score(y_test_actual, lstm_attention_preds)
lstm_cnn_attention_r2 = r2_score(y_test_actual, lstm_cnn_attention_preds)

# Print the results
print(f"LSTM + Attention Model MSE: {lstm_attention_mse:.4f}, MAE: {lstm_attention_mae:.4f}, R²: {lstm_attention_r2:.4f}")
print(f"LSTM + CNN + Attention Model MSE: {lstm_cnn_attention_mse:.4f}, MAE: {lstm_cnn_attention_mae:.4f}, R²: {lstm_cnn_attention_r2:.4f}")

# Plot the predictions of both models against actual values
plt.figure(figsize=(10, 6))
plt.plot(y_test_actual, label='Actual', color='black')
plt.plot(lstm_attention_preds, label='LSTM + Attention Predictions', color='blue', linestyle='--')
plt.plot(lstm_cnn_attention_preds, label='LSTM + CNN + Attention Predictions', color='red', linestyle='--')
plt.title('Model Comparison: LSTM + Attention vs LSTM + CNN + Attention')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import os
import json
from datetime import datetime
from keras_convattention.convattention import ConvAttentionLayer
from keras_convattention.attention import Attention

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create results directory
results_dir = '../Results'
os.makedirs(results_dir, exist_ok=True)

# Load and preprocess data
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, header=0, index_col=0, parse_dates=True)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))

def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 12
X, y = create_dataset(scaled_data, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)

train_size = int(len(X) * 0.8)
val_size = int(len(X) * 0.1)
X_train, X_val, X_test = X[:train_size], X[train_size:train_size+val_size], X[train_size+val_size:]
y_train, y_val, y_test = y[:train_size], y[train_size:train_size+val_size], y[train_size+val_size:]

# Model definitions
def build_lstm_attention_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Attention(units=128, score='bahdanau'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def build_lstm_cnn_attention_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(ConvAttentionLayer(filters=64, kernel_size=3, pool_size=2, units=128, score='bahdanau', input_shape=input_shape))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Model training
lstm_attention_model = build_lstm_attention_model((X_train.shape[1], X_train.shape[2]))
lstm_cnn_attention_model = build_lstm_cnn_attention_model((X_train.shape[1], X_train.shape[2]))

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

lstm_attention_history = lstm_attention_model.fit(X_train, y_train, epochs=100, batch_size=32,
                                                  validation_data=(X_val, y_val), verbose=1,
                                                  callbacks=[early_stopping])
lstm_cnn_attention_history = lstm_cnn_attention_model.fit(X_train, y_train, epochs=100, batch_size=32,
                                                         validation_data=(X_val, y_val), verbose=1,
                                                         callbacks=[early_stopping])

# Predictions
lstm_attention_preds = lstm_attention_model.predict(X_test)
lstm_cnn_attention_preds = lstm_cnn_attention_model.predict(X_test)

# Inverse transform
lstm_attention_preds = scaler.inverse_transform(lstm_attention_preds)
lstm_cnn_attention_preds = scaler.inverse_transform(lstm_cnn_attention_preds)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate metrics
metrics = {
    'LSTM_Attention': {
        'MSE': mean_squared_error(y_test_actual, lstm_attention_preds),
        'MAE': mean_absolute_error(y_test_actual, lstm_attention_preds),
        'R2': r2_score(y_test_actual, lstm_attention_preds)
    },
    'LSTM_CNN_Attention': {
        'MSE': mean_squared_error(y_test_actual, lstm_cnn_attention_preds),
        'MAE': mean_absolute_error(y_test_actual, lstm_cnn_attention_preds),
        'R2': r2_score(y_test_actual, lstm_cnn_attention_preds)
    }
}

# Calculate improvement percentages
improvement = {
    'MSE_improvement_percent': 100 * (metrics['LSTM_Attention']['MSE'] - metrics['LSTM_CNN_Attention']['MSE']) / metrics['LSTM_Attention']['MSE'],
    'MAE_improvement_percent': 100 * (metrics['LSTM_Attention']['MAE'] - metrics['LSTM_CNN_Attention']['MAE']) / metrics['LSTM_Attention']['MAE'],
    'R2_absolute_improvement': metrics['LSTM_CNN_Attention']['R2'] - metrics['LSTM_Attention']['R2']
}

# Save metrics to JSON
results = {
    'timestamp': datetime.now().isoformat(),
    'metrics': metrics,
    'improvement': improvement
}

with open(os.path.join(results_dir, 'performance_metrics.json'), 'w') as f:
    json.dump(results, f, indent=4)

# Plot 1: Original comparison plot
plt.figure(figsize=(10, 6))
plt.plot(y_test_actual, label='Actual', color='black')
plt.plot(lstm_attention_preds, label='LSTM + Attention Predictions', color='blue', linestyle='--')
plt.plot(lstm_cnn_attention_preds, label='LSTM + CNN + Attention Predictions', color='red', linestyle='--')
plt.title('Model Comparison: LSTM + Attention vs LSTM + CNN + Attention')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.legend()
plt.savefig(os.path.join(results_dir, 'original_comparison.png'), dpi=300, bbox_inches='tight')
plt.close()

# Plot 2: Enhanced visualizations
plt.figure(figsize=(18, 12))

# Subplot 1: Validation Loss
plt.subplot(2, 3, 1)
plt.plot(lstm_attention_history.history['val_loss'], label='LSTM+Attention')
plt.plot(lstm_cnn_attention_history.history['val_loss'], label='LSTM+CNN+Attention')
plt.title('Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Subplot 2: Metrics Comparison
metrics_names = ['MSE', 'MAE', 'R²']
attention_scores = [metrics['LSTM_Attention']['MSE'], metrics['LSTM_Attention']['MAE'], metrics['LSTM_Attention']['R2']]
cnn_scores = [metrics['LSTM_CNN_Attention']['MSE'], metrics['LSTM_CNN_Attention']['MAE'], metrics['LSTM_CNN_Attention']['R2']]

plt.subplot(2, 3, 2)
x = np.arange(len(metrics_names))
width = 0.35
plt.bar(x - width/2, attention_scores, width, label='LSTM+Attention')
plt.bar(x + width/2, cnn_scores, width, label='LSTM+CNN+Attention')
plt.xticks(x, metrics_names)
plt.title('Performance Metrics Comparison')
plt.yscale('log')
plt.legend()

# Subplot 3: Error Distribution
attention_errors = y_test_actual - lstm_attention_preds
cnn_errors = y_test_actual - lstm_cnn_attention_preds

plt.subplot(2, 3, 3)
plt.boxplot([attention_errors.reshape(-1), cnn_errors.reshape(-1)], 
            tick_labels=['LSTM+Attention', 'LSTM+CNN+Attention'])
plt.title('Error Distribution Comparison')
plt.ylabel('Prediction Error')

# Subplot 4: Predictions vs Actuals
plt.subplot(2, 3, 4)
max_val = max(y_test_actual.max(), lstm_attention_preds.max(), lstm_cnn_attention_preds.max())
min_val = min(y_test_actual.min(), lstm_attention_preds.min(), lstm_cnn_attention_preds.min())
plt.scatter(y_test_actual, lstm_attention_preds, alpha=0.5, label='LSTM+Attention')
plt.scatter(y_test_actual, lstm_cnn_attention_preds, alpha=0.5, label='LSTM+CNN+Attention')
plt.plot([min_val, max_val], [min_val, max_val], 'k--', label='Ideal')
plt.title('Predictions vs Actual Values')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()

# Subplot 5: Zoomed Comparison
plt.subplot(2, 3, 5)
zoom_size = 20
plt.plot(y_test_actual[-zoom_size:], label='Actual', color='black', marker='o')
plt.plot(lstm_attention_preds[-zoom_size:], label='LSTM+Attention', 
         color='blue', linestyle='--', marker='x')
plt.plot(lstm_cnn_attention_preds[-zoom_size:], label='LSTM+CNN+Attention', 
         color='red', linestyle='--', marker='s')
plt.title('Zoomed Prediction Comparison (Last 20 Points)')
plt.xlabel('Time Step')
plt.ylabel('Passengers')
plt.legend()

# Subplot 6: Relative Improvement
# Subplot 6: Relative Improvement
plt.subplot(2, 3, 6)
improvement_vals = abs(y_test_actual - lstm_cnn_attention_preds) - abs(y_test_actual - lstm_attention_preds)

# Ensure 'color' argument is passed correctly
bar_colors = np.where(improvement_vals < 0, 'green', 'red')  # 'green' or 'red' for bars

# Plot the bars with appropriate colors
plt.bar(range(len(improvement_vals)), improvement_vals, color=bar_colors)  # Use the color array directly
plt.axhline(0, color='black', linestyle='--')
plt.title('Relative Improvement: CNN+Attention vs Attention')
plt.xlabel('Data Points')
plt.ylabel('Error Difference')


plt.tight_layout()
plt.savefig(os.path.join(results_dir, 'detailed_comparison.png'), dpi=300, bbox_inches='tight')
plt.close()

print("Training complete! Results saved to:", results_dir)
```
